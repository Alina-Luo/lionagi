###  documents:str, chunker: str, to_datanode=False

```python
with open('txt_examples/essay.txt') as f:
    essay = f.read()

chunks1 = chunk(essay, 'CharacterTextSplitter', 
                chunker_type='langchain', chunker_kwargs={'chunk_size':100, 'chunk_overlap': 20, 'length_function': len},
                to_datanode=False)
chunks1
```

###  documents:str, chunker: str, to_datanode=True

```python
chunks2 = chunk(essay, 'CharacterTextSplitter', 
                chunker_type='langchain', chunker_kwargs={'chunk_size':100, 'chunk_overlap': 20, 'length_function': len})
chunks2
```

###  documents:List[Document], chunker: Callable, to_datanode=False

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
dnodes3 = load(TextLoader, reader_type=ReaderType.LANGCHAIN, reader_args=[file], to_datanode=False)
chunks3 = chunk(dnodes3, CharacterTextSplitter, 
                chunker_type='langchain', chunker_kwargs={'chunk_size':100, 'chunk_overlap': 20, 'length_function': len},
                to_datanode=False)
chunks3
```

###  documents:List[DataNode], chunker: Callable, to_datanode=True

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
dnodes4 = load(TextLoader, reader_type=ReaderType.LANGCHAIN, reader_args=[file])
chunks4 = chunk(dnodes3, CharacterTextSplitter, 
                chunker_type='langchain', chunker_kwargs={'chunk_size':100, 'chunk_overlap': 20, 'length_function': len},
                documents_convert_func=datanodes_convert)
chunks4
```

# ChunkerType.SELFDEFINED

## Self defined chunker class

must contain function: split()

```python
class demo_chunker:
    def __init__(self, seperator):
        self.seperator = seperator
    
    def split(self, content):
        return content.split(self.seperator)
```

```python
chunks8 = chunk(essay, demo_chunker, 'self_defined', chunker_args=['.'], to_datanode=False)
chunks8
```

## with self_defined to_datanode function

```python
from lionagi.schema.base_schema import DataNode
def to_datanode(content):
    dlist = []
    for c in content:
        dlist.append(DataNode(content=c))
    return dlist
```

```python
chunks9 = chunk(essay, demo_chunker, 'self_defined', chunker_args=['.'], to_datanode=to_datanode)
chunks9
```
